```python
df = pd.read_csv(r'/Users/kanimozhivelusamy/Downloads/DAMG 7275/ProjectP3/company_table.csv')

df.head(100)
```

```python
region_data = df[['RegionID', 'Location', 'TotalEmployee']]

print(region_data)
```

```python
region_data_modified = region_data.rename(columns={'RegionID': 'RegionID', 'Location': 'RegionDescription', 'TotalEmployee': 'TotalWorkForce'})

print(region_data_modified)
```

```python
region_combined = region_data_modified.groupby('RegionID', as_index=False).agg({'RegionDescription': 'first', 'TotalWorkForce': 'sum'})

print(region_combined)
```

```python
row_num = len(region_combined)

region_code = ['RC' + str(np.random.randint(1, 100)).zfill(2) for _ in range(row_num)]

region_map = dict(zip(region_combined['RegionID'], region_code))

region_combined['RegionCode'] = region_combined['RegionID'].map(region_map)

print(region_combined)
```

```python
region_combined.to_csv('region_table.csv', index=False)
```
